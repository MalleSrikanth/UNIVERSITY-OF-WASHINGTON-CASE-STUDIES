```python
def freq(word, doc):
    return doc.count(word)
```

```python
def word_count(doc):
    return len(doc)
```

```python
def tf(word, doc):
    return (freq(word, doc) / float(word_count(doc)))
```

```python
def num_docs_containing(word, list_of_docs):
    count = 0
    for document in list_of_docs:
        if freq(word, document) > 0:
            count += 1
    return 1 + count
```

```python
def tf_idf(word, doc, list_of_docs):
    return (tf(word, doc) * idf(word, list_of_docs))
```

```python

#Compute the frequency for each term.
vocabulary = []
docs = {}
all_tips = []
for tip in (['documment 1', 'documment 2']):
    tokens = tokenizer.tokenize(tip)  #  it changed but it has .text in it  tokens = tokenizer.tokenize(tip.text)
    print('tokens', tokens)
    
    bi_tokens = bigrams(tokens)
    print('bi_tokens' , bi_tokens)
    tri_tokens = trigrams(tokens)
    print('tri_tokens', tri_tokens)
    tokens = [token.lower() for token in tokens if len(token) > 2]
    print('lower_tokens' ,tokens)
    tokens = [token for token in tokens if token not in stopwords]
    print('stopwords_tokens' ,tokens)

    bi_tokens = [' '.join(token).lower() for token in bi_tokens]
    print('lower_bi_tokens' ,bi_tokens)
    
    bi_tokens = [token for token in bi_tokens if token not in stopwords]
    print('stopwords_bi_tokens' ,bi_tokens)

    tri_tokens = [' '.join(token).lower() for token in tri_tokens]
    print('lower_tri_tokens' ,tri_tokens)
    
    tri_tokens = [token for token in tri_tokens if token not in stopwords]
    print('stopwords_tri_tokens' ,tri_tokens)

    final_tokens = []
    final_tokens.extend(tokens)
    final_tokens.extend(bi_tokens)
    final_tokens.extend(tri_tokens)
    print('final_tokens' ,final_tokens)
    
    docs[tip] = {'freq': {}, 'tf': {}, 'idf': {},
                        'tf-idf': {}, 'tokens': []
                }
    print('docs' , docs)
    
    for token in final_tokens:
        #The frequency computed for each tip
        docs[tip]['freq'][token] = freq(token, final_tokens)
        print('doc_freq' ,docs[tip]['freq'][token])
        #The term-frequency (Normalized Frequency)
        docs[tip]['tf'][token] = tf(token, final_tokens)
        print('doc_tf' , docs[tip]['tf'][token])
        docs[tip]['tokens'] = final_tokens
        vocabulary.append(final_tokens)
        
print('final_docs' ,docs)
print('vocabulary' , vocabulary)
```

```python
for doc in docs:
    for token in docs[doc]['tf']:
        print('token _' ,token)
        #The Inverse-Document-Frequency
        docs[doc]['idf'][token] = idf(token, vocabulary)
        print('doc_idf' ,docs[doc]['idf'][token])
        #The tf-idf
        docs[doc]['tf-idf'][token] = tf_idf(token, docs[doc]['tokens'], vocabulary)
        print('doc_tf_idf' , docs[doc]['tf-idf'][token])
```

```python

#Now let's find out the most relevant words by tf-idf.
words = {}

for doc in docs:
    print('doc',doc)
    for token in docs[doc]['tf-idf']:
        print('token' ,token)
        if token not in words:
            words[token] = docs[doc]['tf-idf'][token]
            print('words[token]' ,words[token])
        else:
            if docs[doc]['tf-idf'][token] > words[token]:
                words[token] = docs[doc]['tf-idf'][token]
                print('words[token]',words[token])

    print (doc)
    for token in docs[doc]['tf-idf']:
        print (token, docs[doc]['tf-idf'][token] )
```

```python
from string import punctuation
from nltk.corpus import stopwords
from nltk import word_tokenize
 
stop_words = stopwords.words('english') + list(punctuation)
 
```

```python
def tokenize(text):
    words = word_tokenize(text)
    words = [w.lower() for w in words]
    return [w for w in words if w not in stop_words and not w.isdigit()]
```

```python
# build the vocabulary in one pass
vocabulary = set()
for file_id in reuters.fileids():
    #print(file_id)
    words = tokenize(reuters.raw(file_id))
    #print(words)
    vocabulary.update(words)
```

```python

vocabulary = list(vocabulary)
word_index = {w: idx for idx, w in enumerate(vocabulary)}
print(word_index)
 
VOCABULARY_SIZE = len(vocabulary)
DOCUMENTS_COUNT = len(reuters.fileids())
 
print (VOCABULARY_SIZE, DOCUMENTS_COUNT )     # 10788, 51581
 
```

```python
import numpy as np
 
word_idf = np.zeros(VOCABULARY_SIZE)
for file_id in reuters.fileids():
    words = set(tokenize(reuters.raw(file_id)))
    indexes = [word_index[word] for word in words]
    #print(indexes)
    word_idf[indexes] += 1.0

word_idf = np.log(DOCUMENTS_COUNT / (1 + word_idf).astype(float))
print (word_idf[word_index['deliberations']] )     # 7.49443021503
print (word_idf[word_index['committee']] )        # 3.61286641709
 
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
 
tfidf = TfidfVectorizer(stop_words=stop_words, tokenizer=tokenize, vocabulary=vocabulary)
 
# Fit the TfIdf model
tfidf.fit([reuters.raw(file_id) for file_id in reuters.fileids()])
 
# Transform a document into TfIdf coordinates
X = tfidf.transform([reuters.raw('test/14829')])
 

# Check out some frequencies
print ( X[0, tfidf.vocabulary_['year']] )                  # 0.0562524229373
print (X[0, tfidf.vocabulary_['following']] )             # 0.057140265658
print (X[0, tfidf.vocabulary_['provided']] )              # 0.0689364372666
print (X[0, tfidf.vocabulary_['structural']] )            # 0.0900802810906
print (X[0, tfidf.vocabulary_['japanese']] )              # 0.114492409303
print (X[0, tfidf.vocabulary_['downtrend']] )             # 0.111137191743
 
```
